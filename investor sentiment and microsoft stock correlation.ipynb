```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sp
```

```python
investor = pd.DataFrame.from_csv('investor.csv')
investor = investor.fillna(investor.mean())
investor.tail()
```

```python
msft = pd.DataFrame.from_csv('MSFT.csv')
msft = msft.fillna(msft.mean())
msft.head()
```

```python
week=investor.index 
msftclean=msft.loc[week, :]
msftclean=msftclean.fillna(msft.mean())
close = msftclean.Close
op = msftclean.Open
high = msftclean.High
low = msftclean.Low
```

```python
bulllist = np.array((investor['Bullish'].tolist()))
bearlist = np.array((investor['Bearish'].tolist()))
neutlist = np.array((investor['Neutral'].tolist()))

closelist = np.array(close.tolist())
oplist = np.array(op.tolist())
highlist = np.array(op.tolist())
lowlist = np.array(op.tolist())
```

```python
highbull = sp.pearsonr(highlist, bulllist)
highneut = sp.pearsonr(highlist, neutlist)
highbear = sp.pearsonr(highlist, bearlist)

lowbull = sp.pearsonr(lowlist, bulllist)
lowneut = sp.pearsonr(lowlist, neutlist)
lowbear = sp.pearsonr(lowlist, bearlist)

opbull = sp.pearsonr(oplist, bulllist)
opneut = sp.pearsonr(oplist, neutlist)
opbear = sp.pearsonr(oplist, bearlist)

closebull = sp.pearsonr(closelist, bulllist)
closeneut = sp.pearsonr(closelist, neutlist)
closebear = sp.pearsonr(closelist, bearlist)

print ('Correlation between High stock value and Bullish behaviour is: {0}'.format(highbull))
print ('Correlation between High stock value and Neutral behaviour is: {0}'.format(highneut))
print ('Correlation between High stock value and Bearish behaviour is: {0}'.format(highbear))
print('')
print ('Correlation between Low stock value and Bullish behaviour is: {0}'.format(lowbull))
print ('Correlation between Low stock value and Neutral behaviour is: {0}'.format(lowneut))
print ('Correlation between Low stock value and Bearish behaviour is: {0}'.format(lowbear))
print('')
print ('Correlation between Opening stock value and Bullish behaviour is: {0}'.format(opbull))
print ('Correlation between Opening stock value and Neutral behaviour is: {0}'.format(opneut))
print ('Correlation between Opening stock value and Bearish behaviour is: {0}'.format(opbear))
print('')
print ('Correlation between Closing stock value and Bullish behaviour is: {0}'.format(closebull))
print ('Correlation between Closing stock value and Neutral behaviour is: {0}'.format(closeneut))
print ('Correlation between Closing stock value and Bearish behaviour is: {0}'.format(closebear))
```
