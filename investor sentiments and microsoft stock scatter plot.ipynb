```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
msft = pd.DataFrame.from_csv('MSFT.csv')

investor = pd.DataFrame.from_csv('investor.csv')
bull = investor.Bullish
neut = investor.Neutral
bear = investor.Bearish
investor.tail()
```

```python
msft.tail()
```

```python
week=investor.index 
msft2=msft.loc[week, :]
close = msft2.Close
op = msft2.Open
high = msft2.High
low = msft2.Low
```

```python
plt.figure(figsize=(18,20))
plt.scatter(bull, op, c='r',alpha=0.4, s=50)
plt.scatter(bull, close, c='b',alpha=0.4, s=50)
plt.scatter(bull, high, c='y',alpha=0.4, s=50)
plt.scatter(bull, low, c='g',alpha=0.4, s=50)
plt.legend(fontsize='large')
plt.title('Bullish vs stock')
```

```python
plt.figure(figsize=(18,20))
plt.scatter(neut, op, c='r',alpha=0.4, s=50)
plt.scatter(neut, close, c='b',alpha=0.4, s=50)
plt.scatter(neut, high, c='y',alpha=0.4, s=50)
plt.scatter(neut, low, c='g',alpha=0.4, s=50)
plt.legend(fontsize='large')
plt.title('Neutral vs stock')
```

```python
plt.figure(figsize=(18,20))
plt.scatter(bear, op, c='r',alpha=0.4, s=50)
plt.scatter(bear, close, c='b',alpha=0.4, s=50)
plt.scatter(bear, high, c='y',alpha=0.4, s=50)
plt.scatter(bear, low, c='g',alpha=0.4, s=50)
plt.legend(fontsize='large')
plt.title('Bearish vs stock')
```
