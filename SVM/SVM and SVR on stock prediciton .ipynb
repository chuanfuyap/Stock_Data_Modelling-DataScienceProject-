```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.svm import SVC
import scipy.stats as sp
from sklearn.linear_model import LinearRegression as lr
```

```python
msft = pd.DataFrame.from_csv('MSFT.csv')
msft = msft.interpolate()
investor = pd.DataFrame.from_csv('investor.csv')
investor = investor.fillna(investor.mean())
```

Doing SVRegression first

```python
week=investor.index 
msftclean=msft.loc[week, :]
msftclean=msftclean.interpolate()
```

```python
investor2005 = investor['2005']
investor2005 = investor2005.reindex(index=investor2005.index[::-1])
msft2005 = msft.loc[investor2005.index].interpolate()
op2005 = msft2005.Open
```

```python
plt.scatter(investor2005.Bullish, op2005, color='Blue')
plt.scatter(investor2005.Neutral, op2005, color = 'Red')
plt.scatter(investor2005.Bearish, op2005, color='Green')
```

```python
bulllist = np.array((investor2005['Bullish'].tolist()))
bearlist = np.array((investor2005['Bearish'].tolist()))
neutlist = np.array((investor2005['Neutral'].tolist()))

oplist = np.array(op2005.tolist())

opbull = sp.pearsonr(oplist, bulllist)
opneut = sp.pearsonr(oplist, neutlist)
opbear = sp.pearsonr(oplist, bearlist)

print ('Correlation between Opening stock value and Bullish behaviour is: {0}'.format(opbull))
print ('Correlation between Opening stock value and Neutral behaviour is: {0}'.format(opneut))
print ('Correlation between Opening stock value and Bearish behaviour is: {0}'.format(opbear))
```

It seems in the year 2005 there is positive correlation between bullish behaviour on the Opening stock value for microsoft for year. First, lets try a simple regression Model

```python
bullish2005 = investor2005['Bullish']
bullarray = np.array(bullish2005.tolist())
openarray = np.array(op2005.tolist())
bullarray = bullarray.reshape(bullarray.size,1)
regressionmodel =lr()
regressionmodel.fit(bullarray, openarray)
```

Now lets make a prediction and validate the model

```python
investor2006 = investor['2006']
investor2006 = investor2006.reindex(index=investor2006.index[::-1])
bullish2006 = investor2006['Bullish']
msft2006two = msft.loc[bullish2006.index].interpolate()
open2006two = msft2006two.Open
bullarray2 = np.array(bullish2006.tolist())
bullarray2 = bullarray2.reshape(bullarray2.size,1)
```

```python
pred2006 = regressionmodel.predict(bullarray2)
pred2006 = pd.Series(pred2006, bullish2006.index)
pred2006.plot(color='red')
open2006two.plot(color='blue')
```

```python
pred2006list = np.array(pred2006.tolist())
ori2006list = np.array(open2006two.tolist())

print('Pearson Correlation Coeefiecient: ', sp.pearsonr(ori2006list, pred2006list))

error2 = [((x-y)/y)**2 for x, y in zip(pred2006list, ori2006list)]
print ("Normalised RMSE",np.sqrt(np.sum(error2)))
```

Interestingly eonugh the prediction from SVR using all 3 investor sentiments as features are very close to the predictions with the Linear Regression model using just one feature (which has positive correlation). 

Now Lets try using SVC to classify and predict whether stock will go up or down. First, lets process the data to convert them into +1 and -1. 

```python
lastdayof2004 = investor['2004']
lastdayof2004 = lastdayof2004.index[0]
last2004stock = msft.loc[lastdayof2004].Open
```

```python
UpOrDown2005 = []
previousday=last2004stock
for stock in op2005:    
    if (stock > previousday):
        UpOrDown2005.append(1)
    else:
        UpOrDown2005.append(-1)
    previousday = stock
```

```python
UpOrDown2005 = pd.Series(UpOrDown2005, op2005.index)
```

```python
svcModel = SVC()
```

As investor sentiment information is rather limited, let's make prediction using monthly data, in which the model will predict the day following that month.

```python
year2005month = []
year = '2005-'
for x in range(1,13):
    month = str(x)
    year2005month.append(year+month)
```

```python
results = []
error = []
for month in year2005month:
    try:
        results.append(svcModel.fit(investorSentiments2005[month], UpOrDown2005[month]))
    except ValueError:
        print month, 'cannot be predicted'
        error.append(month)
```

```python
year2005month.remove('2005-6')
year2005month.remove('2005-1')
```

```python
for model, month in zip(results, year2005month):
    print 'Prediction:',model.predict(investorSentiments2005[month].iloc[0:2]), 'Actual:', UpOrDown2005[month].iloc[0:2].tolist()
    
```

For the year 2005, 10 models were fitted, 2 failed (june and december due to lack of variety of classes in those two months), and of these 10 fitted models, only 3 of them were accurate.
